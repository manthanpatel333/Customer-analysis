```python
import re
java['review_text'] = java['review_text'].apply(lambda x: re.sub(r'\d+', '', x))

```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

for index, row in java.iterrows():
    words = row['review_text'].split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    java.at[index, 'review_text'] = ' '.join(filtered_words)
```

```python
java
```

```python

from nltk.stem import WordNetLemmatizer

nltk.download('punkt')

lemmatizer = WordNetLemmatizer()

java['review_text'] = java['review_text'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in nltk.word_tokenize(x)]))
```

```python
# Assuming 'java' is your DataFrame
count_0 = java['recommendation'].eq(0).sum()
count_1 = java['recommendation'].eq(1).sum()

print("Count of 0 in 'recommendation' column: ", count_0)
print("Count of 1 in 'recommendation' column: ", count_1)
```

```python
# Assuming 'review_text' is the column containing your text data
unique_words = set()
for text in java['review_text']:
    words = text.split()  # Splitting the text into words
    unique_words.update(words)

num_unique_words = len(unique_words)
print("Number of unique words in review_text:", num_unique_words)
```

```python
X = java['review_text'].values
y = java['recommendation'].values
```

```python
n_words = 40219

from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=n_words)
```

```python
tokenizer.fit_on_texts(X)
```

```python
dx_tokene = tokenizer.texts_to_sequences(X)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences


X_pad = pad_sequences(dx_tokene, maxlen=1300)# i want to use maximum of all data but it is not possible with becouse it shoes the limitation of memroy


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pad, y, test_size=0.2, stratify=y, random_state=101)
```

```python
# Assuming 'java' is your DataFrame
java['num_tokens_alternative'] = java['review_text'].apply(lambda tokens: len(tokens))

# Convert the result to a NumPy array if needed
num_tokens_alternative = java['num_tokens_alternative'].to_numpy()

# Print the results
print("Number of Tokens using Alternative Logic:")
print(num_tokens_alternative)
```

```python
java
```

```python
dat1 =java['num_tokens_alternative'].max()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Embedding
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

model.add(Embedding(input_dim=num_unique_words,
                    output_dim=300,
                    input_length=dat1,
                    name='embedding_layer'))
model.add(GRU(units=48, return_sequences=True))
model.add(GRU(units=24, return_sequences=True))
model.add(GRU(units=12))
model.add(Dense(1, activation='sigmoid'))
c = Adam(learning_rate=0.003)
model.compile(optimizer= c , loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
pd.Series(y_train).value_counts(normalize=True)
```

```python

class_weights = {0:94, 1:5}

history = model.fit(
    X_train, y_train,
    epochs=3,
    batch_size=256,
    class_weight=class_weights,
    validation_data=(X_test, y_test)
)

evaluate_metrics = model.evaluate(X_test, y_test)
print(f"Loss: {evaluate_metrics[0]}")
print(f"Accuracy: {evaluate_metrics[1]}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score, roc_auc_score

def evaluate_classification_model(model, X, y, set_name):
    # Make predictions
    y_pred = (model.predict(X) >= 0.5).astype("int32")

    # Calculate metrics
    conf_matrix = confusion_matrix(y, y_pred)
    classification_rep = classification_report(y, y_pred)
    accuracy = accuracy_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    roc_auc = roc_auc_score(y, y_pred)

    # Print results
    print(f"Confusion Matrix ({set_name} Set):\n{conf_matrix}")
    print(f"Classification Report ({set_name} Set):\n{classification_rep}")
    print(f"Accuracy ({set_name} Set): {accuracy}")
    print(f"F1 Score ({set_name} Set): {f1}")
    print(f"ROC AUC Score ({set_name} Set): {roc_auc}")

# Evaluate on the training set
evaluate_classification_model(model, X_train, y_train, "Training")

# Evaluate on the test set
evaluate_classification_model(model, X_test, y_test, "Test")
```
